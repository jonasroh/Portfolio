## <font  color = purple> Script

```python
import pandas as pd
from time import sleep
from selenium import webdriver
```

```python
url = 'https://store.steampowered.com/search/?category1=998&os=win&filter=topsellers'
```

```python
wd = webdriver.Chrome()
wd.get(url)
```

```python
# A função "coleta_informacoes" irá coletar todas as informações do jogo.

def coleta_informacoes():
    jogo = {}
    
    # nome do jogo
    nome_do_jogo = wd.find_element_by_id('appHubAppName').get_attribute('innerText')
    jogo['Nome'] = nome_do_jogo

    # preço do jogo
    try:
        wd.find_element_by_class_name('game_purchase_discount_countdown').get_attribute('innerText')
        preco_original = wd.find_element_by_class_name('discount_original_price').get_attribute('innerText')
        preco_final = wd.find_element_by_class_name('discount_final_price').get_attribute('innerText')
    except:
        try:
            preco_final = wd.find_element_by_class_name('game_purchase_price').get_attribute('innerText')
        except:
            preco_final = wd.find_element_by_class_name('discount_final_price').get_attribute('innerText')
    preco_original = preco_final

    jogo["Preço original"] = preco_original
    jogo["Preço final"] = preco_final

    # gêneros
    generos_lista = []
    generos = wd.find_elements_by_xpath('//div[@id="genresAndManufacturer"]/a')
    for cada_elemento in generos:
        generos_lista.append(cada_elemento.get_attribute('innerText'))

    # data de lançamento
    data_de_lancamento = wd.find_element_by_class_name('date').get_attribute('innerText')
    jogo['Data de lançamento'] = data_de_lancamento

    # desenvolvedora
    desenvolvedor = wd.find_element_by_id('developers_list').get_attribute('innerText')
    jogo['Desenvolvedor'] = desenvolvedor

    # distribuidora
    distribuidor = wd.find_element_by_xpath('//div[@class="glance_ctn_responsive_left"]/div/div[2]/a').get_attribute('innerText')
    jogo['Distribuidor'] = distribuidor

    # Idioma
    try:
        traducao = wd.find_element_by_xpath('//div[@class="notice_box_content"]/b').get_attribute('innerText')
    except:
        traducao="Português"

    jogo['Idioma'] = traducao

    # Total de análises e Análise Geral
    analises = wd.find_elements_by_xpath('//div[@class="user_reviews_summary_bar"]/div/span')

    try:
        analise_geral = analises[0].get_attribute('innerText')
    except:
        analise_geral = "Sem análises"

    try:    
        total_de_analises = analises[1].get_attribute('innerText')
    except:
        total_de_analises = 0

    jogo['Análise Geral'] = analise_geral  
    jogo['Total de análises'] = total_de_analises

    return jogo
```

```python
# A função "rolar_pag_e_guardar_jogos" irá fazer uma procura dos jogos a venda na página dos mais vendidas da steam
# e armazenar a url de cada jogo em uma variável.

def rolar_pag_e_guardar_jogos():
    i = 0
    sleep(2)
    while i<10:
        i = i + 1
        wd.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        sleep(4)
        
    todas_as_urls = wd.find_elements_by_xpath('//div[@id = "search_resultsRows"]/a')

    todas_as_urls_lista = []
    for cada_elemento in todas_as_urls:
        todas_as_urls_lista.append(cada_elemento.get_attribute('href'))
        
    return todas_as_urls_lista
```

```python
# A "função restricao_por_idade" irá verificar se ao entrar na página de um jogo a steam irá solicitar o ano de nascimento
# da pessoa. Caso solicite, a função irá preencher automaticamente.

def restricao_por_idade():
    wd.find_element_by_xpath('//select[@id="ageDay"]/option[text()="26"]').click()
    wd.find_element_by_xpath('//select[@id="ageMonth"]/option[text()="dezembro"]').click()
    wd.find_element_by_xpath('//select[@id="ageYear"]/option[text()="1991"]').click()
    wd.find_element_by_class_name('btnv6_blue_hoverfade').click()
    sleep(2)
    
```

```python
todas_as_urls_lista = rolar_pag_e_guardar_jogos()
jogos = [] 
i = 0

while len(jogos) < 500:
    
    wd.execute_script("window.open('%s', '_blank')" %str(todas_as_urls_lista[i]))
    wd.switch_to.window(wd.window_handles[1])
    sleep(1)
    
    try:
        jogos.append(coleta_informacoes())
    except:
        try:
            restricao_por_idade()
            jogos.append(coleta_informacoes())
        except:
            pass
        
    i = i + 1
    wd.close()
    wd.switch_to.window(wd.window_handles[0])
    sleep(1)
```

```python
dataset = pd.DataFrame(jogos)
```
